```python
#Problem 0 - Comments on FIgure From Field
#show original figure
from PIL import Image
image = Image.open('Engler et al Cell 2016 fig.png')
image.show()
```

At first glance the figure is very busy. There are many sub figures and many different types of data protrayed at once. However, closer inspection shows the connections between the different aspects of the figure overall and portrays great meaning. In sub figure A, the grid layout an the color coding help direct the reader's eyes to important relationships. This allows any lay reader to understand which molecular marker is associated with which stem cell type without needing an expert understanding of the field, additionally, the arrows within the images direct the eyes to the important features otherwise easily missed. Sub figure B is organized differently and shows a heat map of acctivated genes in different normalization cases. The side by side comparision of the different stiffness cases allow readers to easily see the differnce in expression levels of the different marker genes in each case. The label size and position for each column however is oddly placed and potentially hard to interpret. The most powerful part of the overall figure is sub figure C. Placed on a traditional axis the message here is immidiately clear,  y changing substrate stiffness you can help direct cell lineage decisions. This message is made all the more clear by the color coding and different marker symbols for the three different stem cell cases explored. The color choices in this sub figure also match the color coding from sub figure A which helps tie these two parts of the figure together and provides a link between the meanings of the two independant sub figures. Also, given that the y-axis in sub figure 3 ranges from 0 to 1, it matches the expression levels depicted by sub figure B and gives context to the graph values in sub figure C.

```python
#Problem 1 - Recreate A Figure of Your Own
#Initialize
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.cm as cm
```

```python
#Load Data and Show Original Fig
SAXS = pd.read_csv('SAXS_data.csv')
image = Image.open('SAXS NanoParticle Fig.png')
image.show()
SAXS
```

```python
#ReCreate the Numeric Parts of the Figure (couldn't figure out how to draw the shapes correctly)
core = SAXS['Core Radius']
shell = SAXS['Shell Radius']
core_err = SAXS['Core Std Dev']
shell_err = SAXS['Shell Std Dev']
tot = SAXS['Tot. Rad']
tot_err = SAXS['Tot. Err']
x=SAXS['Au %']
f, ax = plt.subplots(figsize = (10,6))
width = 2  
core_plot = ax.bar(x, core, width, align='center', alpha= 1, color = 'b', yerr=core_err)
rad_plot = ax.bar(x, tot, width, align='center', alpha = 0.2, color = 'b', yerr=tot_err)


ax.set_xticks(SAXS['Au %'])
ax.set_title('Nanogalvanic Exchange Particle Transformation', fontsize = 22)
ax.set_ylabel('Nanoparticle Radius (nm)', fontsize = 20)
ax.set_xlabel('Nanoparticle Au Content (%) ', fontsize = 20)
minorticks =np.arange(-1, 15, 2)
ax.set_yticks(minorticks,minor = True)

ax.legend((core_plot[0], rad_plot[0]), ('Hollow Core Radius', 'Total Particle Radius'))
```

```python
#Problem 2: reproduce example plot
NYT=pd.read_csv('ny_temps.txt', sep ='\t' )
goog=pd.read_csv('google_data.txt', sep ='\t' )
yahoo=pd.read_csv('yahoo_data.txt', sep ='\t' )

f, ax = plt.subplots(figsize = (10,6))
ax.set_xlim(48988, 55166)
majorticksx=np.arange(49000,55001,1000)
minorticksx=np.arange(48800,55601,200)
ax.set_xticks(majorticksx, minor = False)
ax.set_xticks(minorticksx, minor = True)
ax.set_xlabel('Date (MJD)',fontsize = 16)

minorticksy1=np.arange(-10, 751, 20)
ax.set_yticks(minorticksy1, minor = True)
ax.set_title('New York Temperature, Google, and Yahoo!', fontsize = 20, family='serif', fontweight = 'bold')
pt1 = ax.plot(yahoo['Modified Julian Date'],yahoo['Stock Value'],'b',label = 'Yahoo! Stock Values')
pt2 = ax.plot(goog['Modified Julian Date'],goog['Stock Value'],'purple',label = 'Google Stack Values')
ax.set_ylabel('Value (Dollars)',fontsize = 16)

ax2 = ax.twinx()
pt3 = ax2.plot(NYT['Modified Julian Date'],NYT['Max Temperature'],'r--',label = 'NY Mon. High Temp')
ax2.set_ylim(-150, 100 )
ax2.set_ylabel('Temperature ($^{\circ}$F)',fontsize = 16)
minorticksy2=np.arange(-150, 100, 10)
ax2.set_yticks(minorticksy2,minor = True)

pt = pt1+pt2+pt3
pts = [l.get_label() for l in pt]
ax.legend(pt, pts, loc='center left',frameon = False)

ax.tick_params(axis = 'both', which = 'both', direction = 'in')
ax2.tick_params(axis = 'both', which = 'both', direction = 'in')





```
